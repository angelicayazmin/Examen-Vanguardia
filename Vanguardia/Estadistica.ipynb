```python
from sklearn.datasets import fetch_openml
import pandas as pd

# Base de datos 'adult' desde OpenML
dataset = fetch_openml(name='adult', version=2, as_frame=True)

# datos y etiquetas
data = dataset['data']
target = dataset['target']

# combinar las características con la variable en un solo DataFrame
df = pd.concat([data, target], axis=1)

# descripción del dataset
print("Descripción del dataset:")
print(df.describe(include='all'))

# imprimir las 1ras filas
print("\nPrimeras filas del dataset:")
print(df.head())

# imprimir el tipo de variables
print("\nTipo de variables:")
print(df.dtypes)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_openml
import pandas as pd

# base de datos 'adult' desde OpenML
dataset = fetch_openml(name='adult', version=2, as_frame=True)

# obtener los datos y etiquetas
data = dataset['data']
target = dataset['target']

# combinar las características con la variable objetivo en un solo DataFrame
df = pd.concat([data, target], axis=1)

# obtener los parámetros de todas las variables
print("Parámetros estadísticos de todas las variables:")
print(df.describe(include='all'))

# Gráficas 

# 1. Variables nominales - gráfica de barras para workclass
plt.figure(figsize=(10, 6))
sns.countplot(x='workclass', data=df, order=df['workclass'].value_counts().index)
plt.title('Distribución del Tipo de Trabajo')
plt.xticks(rotation=90)
plt.xlabel('Tipo de Trabajo')
plt.ylabel('Frecuencia')
plt.show()

# 2. Variables ordinales - (ordenado por niveles)
plt.figure(figsize=(10, 6))
education_order = ['Preschool', '1st-4th', '5th-6th', '7th-8th', '9th', '10th', '11th', '12th', 'HS-grad', 'Some-college', 'Assoc-acdm', 'Assoc-voc', 'Bachelors', 'Masters', 'Doctorate', 'Prof-school']
sns.countplot(x='education', data=df, order=education_order)
plt.title('Distribución del Nivel Educativo')
plt.xticks(rotation=90)
plt.xlabel('Nivel Educativo')
plt.ylabel('Frecuencia')
plt.show()

# 3. Variables discretas - (años de educación)
plt.figure(figsize=(8, 6))
sns.histplot(df['education-num'], kde=False, bins=16)
plt.title('Histograma de Años de Educación')
plt.xlabel('Años de Educación')
plt.ylabel('Frecuencia')
plt.show()

# 4. Variables continuas - (edad)
plt.figure(figsize=(8, 6))
sns.boxplot(y='age', data=df)
plt.title('Boxplot de Edad')
plt.ylabel('Edad')
plt.show()

# 5. Variables continuas - (ganancia de capital)
plt.figure(figsize=(8, 6))
sns.histplot(df['capital-gain'], kde=True, bins=30)
plt.title('Histograma de Ganancia de Capital')
plt.xlabel('Ganancia de Capital')
plt.ylabel('Frecuencia')
plt.show()
```
